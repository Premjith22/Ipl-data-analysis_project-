```python
# 🏏 Batsman Analysis
```

```python
#Top 10 Run Scorers in IPL History

top_batsman=deliveries.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False).head(10)
top_batsman
```

```python
#visualization
```

```python
top_batsman.plot(kind="bar",color="tomato", edgecolor="black")
plt.title("Top 10 Run Scorers in IPL History")
plt.xlabel("Batters")
plt.ylabel("Runs")
plt.show()
```

```python
#2 Top 10 Players with Most Sixes

top_sixes = deliveries[deliveries["batsman_runs"] == 6]["batter"].value_counts().head(10)

print(top_sixes)
```

```python
#Visualization
top_sixes.plot(kind="bar")

plt.title("Top Players with Most Sixes")
plt.xlabel("Players")
plt.ylabel("Number of Sixes")

plt.show()
```

```python
#3 Top 10 Players with Most Fours

top_fours= deliveries[deliveries["batsman_runs"]==4]["batter"].value_counts().head(10)
top_fours
```

```python
top_fours.plot(kind="bar")

plt.title("Top Players with Most Fours")
plt.xlabel("Players")
plt.ylabel("Number of Fours")

plt.show()
```

```python
#4 Highest Individual Score
highest_score= deliveries.groupby(["match_id","batter"])["batsman_runs"].sum().sort_values(ascending=False).head(5)
highest_score
```

```python
deliveries.head(1)
```

```python
#5 Which batsman played the most balls in IPL history?

most_balls= deliveries.groupby("batter")["ball"].count().sort_values(ascending =False).head(5)
most_balls
```

```python
most_balls.plot(kind="bar",color="green", edgecolor="black")

plt.title("Players Who Faced Most Balls")
plt.xlabel("Players")
plt.ylabel("Balls Faced")

plt.show()
```

```python
# 🏏 Bowling Analysis
```

```python
#Top 10 Wicket Takers in IPL

wickets= deliveries[deliveries["player_dismissed"].notnull()]
```

```python
top_wickets=wickets["bowler"]. value_counts().head(10)
top_wickets
```

```python
top_wickets.plot(kind="barh")

plt.title("Top 10 Wicket Takers in IPL")
plt.xlabel("Bowlers")
plt.ylabel("Wickets")

plt.show()
```

```python
#Which Bowler Bowled Most Balls?

most_balls= deliveries.groupby("bowler")["ball"].sum().sort_values(ascending=False).head(10)
most_balls
```

```python
most_balls.plot(kind="line",marker="o")

plt.title("Most Balls Bowled in IPL")
plt.xlabel("Bowler")
plt.ylabel("Balls Bowled")

plt.show()
```
